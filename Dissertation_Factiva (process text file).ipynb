```python
'''
Remove stop words from a text file.
Tested on Win7 and Linux Mint 19.1.
'''
import nltk
from nltk.corpus import stopwords
# Get nltk stopword list into a set.
stop_words = set(stopwords.words('english'))

new_stopwords=['tampon','tax', 'ha', 'product', 'wa','woman','new','menstrual', 'english', 'period', 'hygiene','say',
              'school', 'httpstcoea1fxdiznn', 'julissaferreras', 'danismcnally', 'asmgarcia', 'thank','great', 'amp',
               'pls', 'state', 'joannacoles', 'thanks','tamponsforall', 'menstrualequity', 'business']
stpwrd = nltk.corpus.stopwords.words('english')
stpwrd.extend(new_stopwords)
 
# Open and read in a text file.
txt_file = open("Tampon_tax_factiva_final_copy.txt")
txt_line = txt_file.read()
txt_words = txt_line.split()

#make lower case
text1_tokens = []
for t in txt_words:
    if t.isalpha():
        t = t.lower()
        text1_tokens.append(t)
        
#Lemmation
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()

text1_clean = []
for t in text1_tokens:
    t_lem = wordnet_lemmatizer.lemmatize(t)
    text1_clean.append(t_lem)
 
# stopwords found counter.
sw_found = 0
 
# If each word checked is not in stopwords list,
# then append word to a new text file.
for check_word in  text1_clean:
    if not check_word.lower() in stpwrd:
        # Not found on stopword list, so append.
        appendFile = open('Tampon_tax_factiva_cleaned.txt','a')
        appendFile.write(" "+check_word)
        appendFile.close()
    else:
        # It's on the stopword list
        sw_found +=1
        print(check_word)
        
 
print(sw_found,"stop words found and removed")
print("Saved as 'Tampon_tax_factiva_cleaned.txt")
```
